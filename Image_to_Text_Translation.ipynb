```python
from PIL import Image
import pytesseract
import openai
import os
```

```python
# Set the path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
#Load and Preprocess Images
# Function to load images from the dataset, including subfolders
def load_images(image_dir):
    images = []
    for root, dirs, files in os.walk(image_dir):
        for file_name in files:
            if file_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                img_path = os.path.join(root, file_name)
                images.append(img_path)
    return images

# Load images from the dataset
image_dir = 'C:\\Users\\rafar2\\Projects\\Img to Txt\\IIIT5K_coco\\IIIT5K_coco\\images'  # Adjust the path as needed
image_paths = load_images(image_dir)
print(f"Loaded {len(image_paths)} images from {image_dir}")
```

```python
#Extract Text Using Tesseract OCR
# Function to extract text from images
def extract_text(image_path):
    try:
        image = Image.open(image_path)
        text = pytesseract.image_to_string(image, config='--psm 6')
        if not text.strip():  # Check if the text is empty
            return "SORRY can't extract"
        return text
    except Exception as e:
        print(f"Error extracting text from {image_path}: {e}")
        return "Error"
# Initialize lists to categorize images
extracted_text_images = []
non_extracted_text_images = []

# Test OCR on images and categorize
for image_path in image_paths[:500]: # Adjust the number of images as needed
    text = extract_text(image_path)
    if text == "SORRY can't extract" or text == "Error":
        non_extracted_text_images.append(image_path)
    else:
        extracted_text_images.append((image_path, text))
# Calculate percentages
total_images = len(image_paths)
count_no_text = len(non_extracted_text_images)
percentage_no_text = (count_no_text / total_images) * 100
# Print results
print("Images with Extracted Text:")
for image_path, text in extracted_text_images:
    print(f"Image: {image_path}")
    print(f"Extracted Text: {text}\n")

print("Images with No Extracted Text:")
for image_path in non_extracted_text_images:
    print(f"Image: {image_path}")
print(f"\nPercentage of images with no extracted text: {percentage_no_text:.2f}%")
```
